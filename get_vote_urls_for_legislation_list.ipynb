```python
import urllib.request
from bs4 import BeautifulSoup
import re
import json

legislation_list_path = 'data/council_key_legislation_1000sample.json'
with open(legislation_list_path,'r') as infile:
    legislation_records = json.load(infile)
print(legislation_records)
```

```python
def get_vote_url(bill_record_url, target_action):
    base_url = 'https://chicago.legistar.com/'
    
    with urllib.request.urlopen(bill_record_url) as response:
        page = response.read()

        soup = BeautifulSoup(page, 'html.parser')

        evenRowClass = 'rgAltRow'
        oddRowClass = 'rgRow'
        els = soup.select('.'+evenRowClass) + soup.select('.'+oddRowClass)
        for el in els:
            entry_list = el.find_all('td')
            for idx,item in enumerate(entry_list):

                # find actions out of committee -
                if item.text[1:-1]=='City Council':  # pull off book-ending new line characters (warning, is this going to break?)
                    action_by = entry_list[idx].text[1:-1] # City Council  (pull of beginning and trailing new line)
                    action = entry_list[idx+1].text    # Failed to Pass     or     Passed
                    result = entry_list[idx+2].text   # Fail    or    Pass

                    s = entry_list[idx+3].a['onclick'].split('\'')[1]
                    vote_url = base_url + s

                    print("identified bill before the council: ")
                    print(repr(action_by), action, result)
                    
                    if action==target_action:                     
                        return vote_url
                    else:
                        print('action does not match target action')
                        
    print('warning: matching entry not found')
    return 'no vote found'
```

```python

base_url = 'https://chicago.legistar.com/'

for idx,record in enumerate(legislation_records):
    
    action = record['action'] # 'Passed' or 'Failed to Pass'
    bill_record_url = record['legislation_url']
    print('current record: ');
    print (bill_record_url, action)
        
    if action=='Introduced':
        print('no vote has occurred')
    else:
        print('fetching url to voting records...')
        vote_url_string = get_vote_url(bill_record_url, action)
        print(vote_url_string)

        legislation_records[idx]['vote_url'] = vote_url_string
    
print(legislation_records)
    
    
```

```python
target_path = 'data/council_key_legislation_1000sample_voteURLs.json'
with open(target_path, 'w') as outfile:
    json.dump(legislation_records, outfile)
```
