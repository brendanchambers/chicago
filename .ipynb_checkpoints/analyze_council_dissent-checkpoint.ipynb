```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

load_file = 'data/council_votes_dataframe.json'
with open(load_file,'r') as infile:
    data_obj = json.loads(json.load(infile)) # whoops I must have added an extra nesting in the pandas dataframe_to_json_string step

vote_matrix=np.array(data_obj['data'])
votes_dataframe = pd.DataFrame(data=vote_matrix, columns=data_obj['columns'], index=data_obj['index'])
print(votes_dataframe)
    
```
